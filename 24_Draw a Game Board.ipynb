```python
# To label the rows and columns with respective numbers for  rows X columns


rows = int(input("Enter the number of rows : "))
columns = int(input("Enter the number of columns : "))
border = '   ---'
border1 = ' ---'
parts = '|   '
parts1 = '|   '
i = 1
j = 1

# Displaying column numbers
for j in range(1,(columns+1)):
    if j == 1:
        print ('    {}'.format(j),end= ' ')
    elif j >= 10:
        print (' {}'.format(j),end= ' ')
    else:
        print ('  {}'.format(j),end= ' ')

# Creating grid of size (rows X columns)
while i <= rows:
    if i == 1:
        print('\n')
    print(border + (border1 * (columns-1)))
    if i < 10:
        print(" {}".format(i)+ parts + (parts1 * (columns)))
    else:
        print("{}".format(i)+ parts1 + (parts1 * (columns))) 
    i+=1
    
print(border + (border1 * (columns-1)))
          
```
